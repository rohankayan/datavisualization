Importing the libraries , my gloal is to create histogram combining multiple distributions

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
#plt.style.use('seaborn-white')

```

```python
#combined 2 normal distribution , with some distance apart 
combinednormaldis = np.concatenate((np.random.randn(1000) , 20 +np.random.randn(500)))
```

```python
#Histogram with Freedman and Diaconisâ€™s (1981) rule
hist_fd= plt.hist(combinednormaldis,bins=100,cumulative=True)
plt.title("Histogram with 'FD' bins on combiner normal distribution")
```

Not able to highlight the normal pattern of the data 

```python
#Histogram with Sturges
hist_auto= plt.hist(combinednormaldis,bins='sturges')
plt.title("Histogram with 'sturges' bins on combiner normal distribution")
```

Looks like a better choice for the bin size

```python

plt.hist(combinednormaldis,bins='sqrt')
plt.title("Histogram with 'sqrt' bins on combiner normal distribution")
```

```python
uniform = np.random.random(100000)

plt.hist(uniform, bins='sturges')

plt.title("Histogram with 'sturges' bins and uniform data large data")

plt.show()
```
